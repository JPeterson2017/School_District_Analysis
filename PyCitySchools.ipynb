```python
import pandas as pd
```

```python
# Files to Load (You can also use os.path.join("Folder name","file name") to load files with unknown location 
# If you do this, import the os directory in above step)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read  school data file and store it in pandas dataframe - the read_csv function automatically converts it to a DF
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head()
```

```python
# Read student file and store into pandas data frame
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```
